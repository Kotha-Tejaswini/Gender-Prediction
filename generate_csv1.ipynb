```python
import numpy as np
import cv2
import os
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
dir=r"C:\test"
x=["b","g"]
shape=(28,28)
```

```python
test_data=[]
test_directory=dir
for i in x:
    path=os.path.join(test_directory,i)
    for img in os.listdir(path): 
        array_image=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        array_image=cv2.resize(array_image,shape)
        array_image=array_image.flatten()/255
        array_image=np.append(array_image,x.index(i))
        test_data.append(array_image)
dfm=pd.DataFrame(test_data)
dfm
```

```python
dfm.to_csv("D:\dataset1.csv",index=False)
```
